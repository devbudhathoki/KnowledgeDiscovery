```python
# This function open the file in read mode and returns the content of file
# @ params: filename  name of input file
#@ returns: lists filecontent of file
def openfile(filename):
    file=open(filename,"r")
    data=file.read()
    return data
```

```python
#This function reads the file content from openfile() and list of data in different lines of files
#@params: file file content of file
#@return: list of data in different lines of files
def filecontent(file):
    lines=file.split("\n")
    #print(lines)
    lists=[]
    for line in lines:
        lists.append(line.split(","))    
    #print(lists)
    return lists
```

```python
def disease_symptoms(lists,disease):
    disease1=[]
    #print(D)
    #print (lists[0:1])
    
    
    for each in lists[1:len(lists)-1]:
        #print(each[disease])
        if (float(each[disease])!=0.0):
            #print(lists[0][disease])
            #print(each[0])
            
            disease1.append(each[0])
    #print(dictionary)
    
    return disease1
```

```python
# write the lists into csv file
#@params: lists lists of data 
#@params: filename output filename 

def writefile(lists,filename):
    import csv
    file=open(filename,"w")
    #writer = csv.DictWriter(file)
    writer=csv.writer(file)
    writer.writerows(lists)
    
```

```python
#This function conversts the comma delimited csv data into matrix using numpy and returns transpose of matrix
#@params: filename Input csv filename
#@params: result.T transpose matrix of Inputfile
def convert_matrix(filename):
    import csv
    import numpy
    reader = csv.reader(open(filename, "r"), delimiter=",")
    x = list(reader)
    #print(x[1:])
    result = numpy.array(x[1:]).astype("str")
    #print(result)
    #print(result.T)
    return (result.T)
    
```

```python
# create the list of Symptoms for a diagnosis from matrix
def make_items(filename,lists,data):
    final1=[]
    for i in range(1,data.shape[1]): 
        disease=[]
        #print(lists[0][i])
        #disease.append(lists[0][i])
        disease=disease_symptoms(lists,i)
        #disease.append(lists[0][i])        
        final1.append(disease)
       # print((disease))
    writefile(final1,filename)
    
    
```

```python
# create list of Symptoms and its respective diagnosis from matrix
def make_basket(filename,lists,data):
    final=[]
    for i in range(1,data.shape[1]): 
        disease=[]
        #print(lists[0][i])
        #disease.append(lists[0][i])
        disease=disease_symptoms(lists,i)
        disease.append(lists[0][i])        
        final.append(disease)
       # print((disease))
    
    writefile(final,filename)
   
```

```python

def create_basket_item(basket_filename,item_filename,sourcefile):
    import pandas as pd
    import csv
    file=sourcefile
    #print(sd_diff.head())
    data=openfile(file)
    lists= filecontent(data)
    df = pd.read_csv(file)
    
    #print(df.columns[1])
    data=df.values
    #print(data)
   # print(data)
    #print(data.shape[1])
   # print(type(''.join(lists[0][1])))
    #print(len(lists)) 
    
    make_items(item_filename,lists,data)
    make_basket(basket_filename,lists,data)
    
```
